Align traces

```julia
reset(trs) # note, this resets the conditional averager, it doesn't pop the sample passes

# selecting the reference pattern in the first races
referencestart = 5800
referenceend = referencestart + 400
reference = trs[1][2][referencestart:referenceend]

# in the search of alignemt, traces will be shifted by max this amount of samples
maxShift = 500

# the rejection threshold
corvalMin = 0.0

# create the alignment engine
alignstate = CorrelationAlignFFT(reference, referencestart, maxShift)

# add the alignment as a sample processing pass
# the end of the next line is somewhat complex I must say :) But it's an experimental toolbox, yay!
addSamplePass(trs, x -> ((shift,corval) = correlationAlign(x, alignstate); corval > corvalMin ? circshift(x, shift) : Vector{eltype(x)}(0)))

# execute the alignment pass by reading the traces
#((data,samples),eof) = readTraces(trs, 1:300)
# remove the alignment apss
#popSamplePass(trs)
```

```julia
# execute the alignment pass by reading the traces
((data,samples),eof) = readTraces(trs, 1:95)
plot(samples[1:95,5000:6000]',linewidth=.3); # note the transpose when plotting multiple traces
```

```julia
# set the parameters
numberOfSamples = size(trs[1][2])[1]
dataSpace = size(trs[1][1])[1]
sampleType = Float32;

# create and save the trs
trs_aligned = InspectorTrace("rhme3-itskindofmagic-aligned.trs", dataSpace, sampleType, numberOfSamples)
for t in 1:numberOfTraces
  trs_aligned[t] = trs[t]
end
close(trs_aligned)
```

## CPA attack

```julia
reset(trs)

params = DpaAttack(AesSboxAttack(), IncrementalCPA())
params.dataOffset = 1       # plaintext starts from byte 1
params.attack.direction = FORWARD
params.analysis.leakages = [HW()]
numberOfTraces = length(trs);

setPostProcessor(trs, IncrementalCorrelation())
rankData = sca(trs, params, 1, numberOfTraces)
round1_key = getKey(params,rankData)
```

```julia
# process input data according to 2nd round
function Round2DataPass(data::Array, roundkey::Array)
    state = reshape(data[1:16], (4, 4))
    key = reshape(roundkey, (4, 4))
    state = Aes.AddRoundKey(state, key)
    state = Aes.SubBytes(state)
    state = Aes.ShiftRows(state)
    state = Aes.MixColumns(state)
    return state[1:16]
end
```

```julia
# add data transformation pass
addDataPass(trs, x::Array -> Round2DataPass(x, round1_key))
# attack 2nd round
params = DpaAttack(AesSboxAttack(), IncrementalCPA())
params.dataOffset = 1
params.attack.direction = FORWARD
params.analysis.leakages = [HW()]
numberOfTraces = length(trs);

setPostProcessor(trs, IncrementalCorrelation())
rankData = sca(trs, params, 1, numberOfTraces)
popDataPass(trs)
#
round2_key = getKey(params, rankData)
```

```julia
key = KeyExpansionBackwards(round2_key, 1, 4)[1:16]
```
